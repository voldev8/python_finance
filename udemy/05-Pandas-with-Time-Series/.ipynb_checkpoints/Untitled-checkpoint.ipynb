```python
df = pd.read_csv('time_data/walmart_stock.csv') #index_col='Date' parse_dates=True   faster way
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info() #now a datetime object
```

```python
#df['Date'] = df['Date'].apply(pd_to_datetime)
```

```python
df.set_index('Date', inplace=True)
```

```python
df
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='BQ').mean()
```

```python
df.resample(rule='A').max()
```

```python
def first_day(entry):
    return entry[0]
```

```python
df.resample('A').apply(first_day)
```

```python
df['Close'].resample('Q').mean().plot(kind='bar', figsize=(16,6))
```

```python
#rolling and expanding (rolling = moving avg)
```

```python
df.head(14)
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df.rolling(7).mean().head(14) #7 day avg first 6 is blank because  it starts averiging at 7th day
```

```python
df['Open'].plot()
df.rolling(window=90).mean()['Close'].plot(figsize=(16,6))
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA','Close']].plot(figsize=(16,6),color=['r','b'])
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
```

```python
#bollinger bands
```

```python
#close 20 ma
df["Close: 20 Day Mean"] = df['Close'].rolling(20).mean()
#upper band = 20ma + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df["Close"].rolling(20).std())
#lower band = 20ma - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df["Close"].rolling(20).std())
#close
df[['Close',"Close: 20 Day Mean",'Upper','Lower']].plot(figsize=(16,6))
```

```python
df[['Close',"Close: 20 Day Mean",'Upper','Lower']].tail(200).plot(figsize=(16,6))
```

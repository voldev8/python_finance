```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#functional method
plt.plot(x,y)
#plt.show() if you re not in jupyter //it is like print()

plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) #number of rows, columns, plot number
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#Object-Oriented method
fig = plt.figure()

axes = fig.add_axes([.1, .1, .8, .8]) #left, bottom, width, height percenteges

axes.plot(x,y)
axes.set_xlabel('X label') 
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([.1, .1, .8, .8])
axes2 = fig.add_axes([.2, .5, .4, .3])

axes1.plot(x,y,'g')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'r')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)

# axes.plot(x,y)
plt.tight_layout() #fixes the overlap
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# for current_ax in axes:
#     current_ax.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

```python
fig = plt.figure(figsize=(3,2),dpi=200)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x Squared')
ax.plot(x, x**3, label='x Cubed')

ax.set_title('Title')
ax.set_xlabel('label x')
ax.set_ylabel('label y')
#adding legends
ax.legend()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4),dpi=200)

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=300)
```

```python
fig = plt.figure(dpi=150)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color="#FF8C00", linewidth=3, linestyle="-", marker='o') #lw ls ==> short

ax.set_xlim([0,3])
```

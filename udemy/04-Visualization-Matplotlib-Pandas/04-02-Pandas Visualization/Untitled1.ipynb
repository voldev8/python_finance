```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
mcdon = pd.read_csv('mcdonalds.csv', index_col='Date',parse_dates=True)
```

```python
mcdon.head()
```

```python
mcdon['Adj. Volume'].plot(figsize=(14,4))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=(20,55),figsize=(14,4))
```

```python
from matplotlib import dates
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock,'-')

#locating
ax.xaxis.set_major_locator(dates.MonthLocator())
#formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

fig.autofmt_xdate()
plt.tight_layout()
```
